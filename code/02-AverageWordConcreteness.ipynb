```python
def avg_word_concreteness(text):
    n = 0
    sum_wc = 0
    
    if(type(text) != str):
        return 0
    
    tokens = text.split()
    for token in tokens:
        #Remove special characters
        token = ''.join(e for e in token if e.isalnum()).lower()
        conc_entry = word_concreteness.loc[word_concreteness['Word'] == token]
        
        if(not conc_entry.empty):
            n += 1
            sum_wc += conc_entry['Conc.M'].values[0]
            #print(token + " score:" + str(conc_entry['Conc.M']))
            #print("Sum score:" + str(sum_wc) + " n:" + str(n) + "\n")
    
    
    return sum_wc/n if n>0 else 0
```

```python
avg_concreteness = [avg_word_concreteness(x) for x in df['Response.x']]
```

```python
df["Concrete.Avg"] = avg_concreteness

print(df)
```

```python
def high_concreteness_count(text, threshold):
    n = 0
    
    if(type(text) != str):
        return 0
    
    tokens = text.split()
    for token in tokens:
        #Remove special characters
        token = ''.join(e for e in token if e.isalnum()).lower()
        conc_entry = word_concreteness.loc[word_concreteness['Word'] == token]
        
        if(not conc_entry.empty and conc_entry['Conc.M'].values[0] > threshold):
            n += 1
    
    return n
```

```python
high_concreteness = [high_concreteness_count(x, 3) for x in df['Response.x']]
df["Concrete.HighCount"] = high_concreteness

print(df)
```

```python
def high_concreteness_density(text, threshold):
    count = 0
    n = 0
    
    if(type(text) != str):
        return 0
    
    tokens = text.split()
    for token in tokens:
        #Remove special characters
        token = ''.join(e for e in token if e.isalnum()).lower()
        conc_entry = word_concreteness.loc[word_concreteness['Word'] == token]
        
        n += 1
        if(not conc_entry.empty and conc_entry['Conc.M'].values[0] > threshold):
            count += 1
    
    return count/n
```

```python
df["Concrete.HighDensity"] = [high_concreteness_density(x, 3) for x in df['Response.x']]
df["Concrete.ExtremeCount"] = [high_concreteness_count(x, 4) for x in df['Response.x']]
df["Concrete.ExtremeDensity"] = [high_concreteness_density(x, 4) for x in df['Response.x']]

print(df)
```

```python
def bigram_high_concreteness_count(text, threshold):
    n = 0
    last_word = ''
    
    if(type(text) != str):
        return 0
    
    tokens = text.split()
    for token in tokens:
        #Remove special characters
        token = ''.join(e for e in token if e.isalnum()).lower()
        
        if(len(last_word)==0):
            last_word = token
            continue
            
        bigram = last_word + " " + token
        conc_entry = bigram_concreteness.loc[bigram_concreteness['Word'] == bigram]
        
        last_word = token
        
        if(not conc_entry.empty and conc_entry['Conc.M'].values[0] > threshold):
            n += 1
    
    return n
```

```python
df["Concrete.BigramHighCount"] = [bigram_high_concreteness_count(x, 3) for x in df['Response.x']]

print(df)
```
