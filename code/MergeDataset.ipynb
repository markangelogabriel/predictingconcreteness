```
import pandas as pd

#Load datasets
df1 = pd.read_csv('CAREStudy_ProbSolv_NLP_Rated_18Jan2021.csv')
df2 = pd.read_csv('SocDist_AllSurveys_Knitted_Clean_v06.csv')
df3 = pd.read_csv('SocDist_ForAnalysis.csv')
```

```
df1.head()
```

```
# Make sure key column is equivalent by setting emails to lowercase
df1['Study.ID'] = df1['Study.ID'].str.lower()
df2['Study.ID'] = df2['Study.ID'].str.lower()
df3['Study.ID'] = df3['Study.ID'].str.lower()

df1.head()
```

```
# Merge full dataset
df_full = df1.merge(df2, on='Study.ID')
df_full.head()
```

```
#Save merged full dataset
df_full.to_csv('SocDist_AllSurveys_Merged_19012021.csv', index=False)
```

```
# Merge short dataset
df_short = df1.merge(df3, on='Study.ID')
df_short.head()
```

```
#Save merged short dataset
df_short.to_csv('SocDist_ForAnalysis_Merged_19012021.csv', index=False)
```
